#### Importing dataset

```python
df=pd.read_csv("/Users/ittismita/Downloads/movies.csv")
```

```python
df.head()
```

#### looking for missing values

```python
#method 1
#percentage mean of missing values in each column
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print("{}-{}%".format(col,pct_missing))
```

```python
#method 2
#number of missing values in each column
df.isnull().sum()
```

```python
df.mean()
```

```python
#as we have null values so now we will replace those null values with 0.
df=df.fillna(0)
```

```python
df.isnull().sum()
```

```python
#looking at the datatypes of our columns
df.dtypes
```

```python
#changing datatype of some columns
df["votes"]=df["votes"].astype('int64')
df["budget"]=df["budget"].astype('int64')
df["gross"]=df["gross"].astype('int64')
df.head()
```

#### formatting date columns

```python
df_split=df["released"].str.split('(')
released_date=df_split.str[0]
released_location=df_split.str[1].str.strip(')')
```

```python
df["released_date"]=released_date
df["released_location"]=released_location
df["released_date"]=pd.to_datetime(df["released_date"])
df.head()
```

```python
df=df.drop(["released"],axis=1)
```

```python
df.head()
```

```python
df["Year"]=df["released_date"].astype("str").str[0:4]
df=df.drop(["year"],axis=1)
```

```python
df.head()
```

```python
#sorting the values by gross
df.sort_values(by="gross",inplace=False,ascending=False).head()
```

#### Plotting correlations

```python
plt.scatter(x=df['budget'],y=df['gross'],c='teal')
plt.title("Budget vs Gross")
plt.xlabel("Budget of the Movie")
plt.ylabel("Gross")
plt.show()
```

```python
sns.regplot(data=df,x=df['budget'],y=df['gross'],scatter_kws={"color":"palevioletred"},line_kws={"color":"purple"})
plt.title("Budget vs Gross")
plt.xlabel("Budget of the Movie")
plt.ylabel("Gross")
```

```python
sns.regplot(data=df,x=df['votes'],y=df['gross'],scatter_kws={"color":"burlywood"},line_kws={"color":"darkorange"})
plt.title("Votes vs Gross")
plt.xlabel("Votes for the Movie")
plt.ylabel("Gross of the Movie")
```

```python
df.corr(method="pearson") #other methods= kendall, spearman...default is pearson
```

```python
#visualizing the above correlation matrix
corr_matrix=df.corr(method="pearson")
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation matrix For Numeric Features")

plt.show()
```
